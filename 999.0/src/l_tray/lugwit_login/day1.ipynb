```python
#连接数据库并创建连接
import pymysql              # 导入 pymysql 模块 
import pandas as pd     qq爱  

# 数据库连接信息
host = 'localhost'  # 数据库服务器地址
user = 'root'  # 数据库用户名
password = '123456'  # 数据库密码

9998
# 建立数据库连接
connection = pymysql.connect(host=host, user=user, password=password)

# 使用 connection 创建一个 cursor 对象
cursor = connection.cursor()
```

```python
# 设置当前的数据库
cursor.execute("USE lugwit_tool")
cursor.execute("SELECT DATABASE()")
SELECT_DATABASE=cursor.fetchone()
print (SELECT_DATABASE)
```

```python
# 创建表
sql = """CREATE TABLE IF NOT EXISTS `users` (
  `id` int(11) NOT NULL AUTO_INCREMENT,
  `username` varchar(50) NOT NULL,
  `email` varchar(100) NOT NULL,
  `password` varchar(100) NOT NULL,
  PRIMARY KEY (`id`)
) ENGINE=InnoDB DEFAULT CHARSET=utf8"""
cursor.execute(sql)
```

```python
# 修改表结构
alter_table_query = """
ALTER TABLE users
ADD UNIQUE (username),
ADD UNIQUE (email);
"""
cursor.execute(alter_table_query)

# 提交更改
connection.commit()
```

```python
# 添加加密的密码
import bcrypt
sql = """INSERT INTO `users` (`username`, `email`, `password`) VALUES (%s, %s, %s)"""
# 用户的原始密码
password = "password123"
# 生成盐
salt = bcrypt.gensalt()
print (salt)
# # 哈希密码
hashed_password = bcrypt.hashpw(password.encode(), salt)
print (hashed_password)
cursor.execute(sql, ('John', 'john@example.com', hashed_password))
# 提交更改
connection.commit()
```

```python
import bcrypt

# 假设 user_input 是用户尝试登录时输入的密码
user_input = "password1234"

# 从数据库获取哈希密码
sql = "SELECT password FROM users WHERE username = %s"
value = ('John',)
cursor.execute(sql, value)
hashed_password_from_db = cursor.fetchone()[-1]
print (hashed_password_from_db)

# 比较用户输入的密码与数据库中的哈希密码
if bcrypt.checkpw(user_input.encode(), hashed_password_from_db.encode()):
    print("Password is correct")
else:
    print("Password is incorrect")
```
